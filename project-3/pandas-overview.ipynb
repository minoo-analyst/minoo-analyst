### Join

```python
autodf2 = autodf[:20].copy(deep=True)
autodf2
```

```python
autodf2.reset_index(inplace=True)
autodf2
```

```python
fillerdf = pd.DataFrame({
    "carName"     : autodf2.carName,
    "Zeros"       : np.zeros(len(autodf2)),
    "Ones"        : np.ones(len(autodf2))
})
fillerdf
```

```python
expanded_df = pd.merge(autodf2, fillerdf, on="carName")
expanded_df
```
